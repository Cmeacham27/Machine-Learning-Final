```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('US Cancer Death Rates by State Machine.csv')
df = df.rename(columns={"Unnamed: 0": "State"})
year_cols = ['2014_Rate','2015_Rate','2016_Rate','2017_Rate', '2018_Rate','2019_Rate','2020_Rate','2021_Rate', '2022_Rate','2023_Rate']
avg_rates = df[year_cols].mean()
years = [col.split("_")[0] for col in year_cols]

plt.figure(figsize=(10,5))
plt.plot(years, avg_rates.values, marker='o')
for i, value in enumerate(avg_rates.values):
    plt.text(i, value + 0.2, f"{value:.1f}", ha='center')

plt.title("Average U.S. Cancer Death Rate (2014–2023)")
plt.xlabel("Year")
plt.ylabel("Average Death Rate (per 100,000)")
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```
