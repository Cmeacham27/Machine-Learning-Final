# 1. ReadMe
This project analyzes U.S. cancer death rates by state from 2014–2023 to find trends, patterns, and build models to try and predict future results. The dataset includes yearly cancer mortality rates, which were used to see how rates change over time and how accurately future rates can be predicted.
## Problem Statement & Goals
The main goal of this project was to determine: Whether cancer death rates follow a predictable trend over time. How accurately we can predict future cancer death rates using historical data. Whether meaningful patterns exist across states and regions
## Overall Approach
The project followed a structured data science pipeline: Data collection and cleaning Exploratory Data Analysis Feature scaling and preprocessing Predictive modeling (regression + classification) Clustering and dimensionality reduction (K-Means + PCA) Model evaluation and interpretation
Multiple models were tested, including: Linear Regression Decision Tree Regressor Support Vector Machines (SVR & SVC) Random Forest Neural Networks (MLP)
## Project Sections
RAW_DATA: Data source and acquisition DATA: Cleaning, preprocessing, and visualization ANALYSIS: Modeling and technical evaluation CONCLUSIONS: Key findings and insights
## Summary
Through this project, I learned that cancer death rates in the U.S. are very stable and follow a consistent downward trend. This makes them very helpful for making predictive modeling. I also learned that while total deaths may increase, death rates can still decline, which provides a more accurate measure of health outcomes. This is because population can grow throughout the year, but the rate can stay the same, or go down. Additionally, different models vary in effectiveness, with neural networks creating more complex patterns than simpler models.

# 4. Analysis
## Models Used
Regression Models:
•	Linear Regression 
•	Decision Tree Regressor 
•	Support Vector Regressor (SVR) 
•	Random Forest Regressor 
•	Neural Network (MLPRegressor) 
Classification Model:
•	Support Vector Classifier (SVC) for risk categories 
Clustering:
•	K-Means (k=3) 
Dimensionality Reduction:
•	PCA (2 components) 
## Model Performance
•	Linear Regression: Strong linear fit (high R²) 
•	Decision Tree: Struggled with high-risk predictions 
•	SVM: 
o	Regression performed well 
o	Classification achieved high accuracy (8/10 correct) 
•	Random Forest: Good performance but slightly higher error 
•	Neural Network: Best performance with lowest MSE (~4.9M) 
## Evaluation Metrics
•	Mean Squared Error (MSE) 
•	R² Score (~0.88 for regression) 
•	Confusion Matrix (classification accuracy) 
## What Worked Well
•	Historical trends strongly predicted future outcomes 
•	Linear relationships made regression highly effective 
•	PCA captured ~97.9% of variance in just 2 components 
•	SVM classification performed reliably across categories 
## What Did Not Work Well
•	Decision Tree struggled with high-risk classification 
•	Some models were sensitive to scaling and preprocessing 
•	Small dataset limited model generalization Challenges & Solutions
Challenges:
•	Deciding between predicting by state vs national trends 
•	Handling classification vs regression approaches 
•	Interpreting conflicting signals (rate vs total deaths) 
Solutions:
•	Focused on rate (per 100,000) instead of total deaths 
•	Used multiple models to compare performance 
•	Applied scaling and PCA for better structure 
## Limitations
•	Limited dataset size (only ~50 states) 
•	No external factors (e.g., healthcare access, demographics) 
•	Assumes trends remain consistent over time
## Possible Improvements
•	Include socioeconomic or healthcare variables 
•	Use time-series models (ARIMA, LSTM) 
•	Perform cross-validation for stronger evaluation 
•	Increase dataset size (county-level data)

# 5. Conclusions
## Key Findings
•	Cancer death rates are steadily declining across the U.S. 
•	Trends are highly consistent year-to-year 
•	Historical data is a strong predictor of future outcomes 
•	Neural networks capture complex patterns best 
## Insights
•	Death rate is a more reliable metric than total deaths 
•	Population growth can mask improvements in mortality 
•	Larger states form distinct clusters due to scale effects 
•	Outliers represent states needing targeted healthcare strategies 
## Story of the Results
This project shows that cancer mortality in the U.S. follows a stable and predictable pattern. While raw death counts may increase due to population growth, the actual risk of dying from cancer is decreasing. By applying machine learning models, especially neural networks and SVMs, we can accurately forecast future death rates and categorize states by risk level. Clustering and PCA further revealed that most variation in the data is driven by population size and mortality intensity, helping simplify complex patterns into understandable insights.
